```python
import pandas as pd
import numpy as np

# -------------------------------
# FILE NAMES
# -------------------------------
input_csv = "SearchMedicineData.csv"
output_csv = "SearchMedicineData_updated.csv"

# -------------------------------
# CONFIG
# -------------------------------
random_dates = ["31/01/2026", "03/03/2026"]
UPDATE_COUNT = 30000
DATE_COLUMN = "expirydate"

# -------------------------------
# READ CSV
# -------------------------------
df = pd.read_csv(input_csv)

# -------------------------------
# RANDOM ROW INDEX SELECTION
# -------------------------------
random_indices = np.random.choice(
    df.index,
    size=UPDATE_COUNT,
    replace=True  # allow same row multiple times
)

# -------------------------------
# ASSIGN RANDOM DATES
# -------------------------------
df.loc[random_indices, DATE_COLUMN] = np.random.choice(
    random_dates,
    size=UPDATE_COUNT
)

# -------------------------------
# SAVE UPDATED CSV
# -------------------------------
df.to_csv(output_csv, index=False)

print("‚úÖ CSV updated successfully!")
print("üìÖ Random dates added:", random_dates)
print("üîÅ Total updates:", UPDATE_COUNT)
print("üìÑ Output file:", output_csv)
```

```python
import pandas as pd
import numpy as np

def remove_duplicates_and_save(input_csv, output_csv):
    """
    Reads a CSV file, removes duplicate rows, and saves the unique data to a new CSV.

    Args:
        input_csv (str): The path to the original CSV file.
        output_csv (str): The path where the new, cleaned CSV file will be saved.
    """
    try:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(input_csv)
        print(f"Original DataFrame shape: {df.shape}")

        # Remove duplicate rows, keeping the first occurrence by default
        # The 'subset' parameter can be used to check for duplicates in specific columns only
        df_cleaned = df.drop_duplicates(keep='first')

        print(f"Cleaned DataFrame shape (after removing duplicates): {df_cleaned.shape}")

        # Save the cleaned DataFrame to a new CSV file
        df_cleaned.to_csv(output_csv, index=False) # index=False prevents writing the DataFrame index as a column
        print(f"Successfully saved unique data to {output_csv}")

    except FileNotFoundError:
        print(f"Error: The file '{input_csv}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- Example Usage ---
# Replace 'your_input_file.csv' with the name of your original file.
# Replace 'your_output_file.csv' with the desired name for the new file (e.g., 'cleaned_data_20mb.csv').
input_filename = 'SearchMedicineData.csv'
output_filename = 'Data.csv'

remove_duplicates_and_save(input_filename, output_filename)
```
